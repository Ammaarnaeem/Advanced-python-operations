### Different operations on image through python
#### Image denoising using cv2 library

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn')

# denoising image
image = cv2.imread('noisy.jpeg')
dst = cv2.fastNlMeansDenoisingColored(image, None, 9, 9, 6, 14)

# saving denoised image
#cv2.imwrite('denoised_image.jpg', dst)

# plotting noised and denoised imag
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Before Denoising')
axs[1].imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
axs[1].set_title('After Fast Means Denoising')
plt.show()
```

#### Creating qr code from image

```python
# producing qr code for the image
import qrcode
from PIL import Image

# Example: Generate a QR code
qr = qrcode.QRCode(version=1, error_correction=qrcode.constants.ERROR_CORRECT_L, box_size=10, border=4)
qr.add_data('itach.jpg')
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white")
img.save('qrcode.png')
```
